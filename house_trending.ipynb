```python
M2_2020_1_df = dsf_util.get_dsf_building_df("./2020/2_1.xls", 2020,2)
M2_2020_2_df = dsf_util.get_dsf_building_df("./2020/2_2.xls", 2020,2)
M2_2020_df = M2_2020_1_df.append(M2_2020_2_df)
M2_2019_1_df = dsf_util.get_dsf_building_df("./2019/2_1.xls", 2019,2)
M2_2019_2_df = dsf_util.get_dsf_building_df("./2019/2_2.xls", 2019,2)
M2_2019_df = M2_2019_1_df.append(M2_2019_2_df)
```

```python
monthly_excel_file = [ '1_1.xls', '1_2.xls', '2_1.xls', '2_2.xls', '3_1.xls', '3_2.xls', '4_1.xls', 
                       '4_2.xls', '5_1.xls', '5_2.xls', '6_1.xls', '6_2.xls', '7_1.xls', '7_2.xls',
                       '8_1.xls', '8_2.xls', '9_1.xls', '9_2.xls', '10_1.xls', '10_2.xls', '11_1.xls', '11_2.xls', 
                       '12_1.xls', '12_2.xls'
                     ]
```

```python
df_2019 = pd.DataFrame({}, columns=['Year','Month','Area','Building_Name','Volume','Price/㎡'])
```

```python
for file in monthly_excel_file:
    idx = file.index("_")
    month = file[0:2] if idx > 1 else file[0]
    month = int(month)
    df_2019 = df_2019.append(dsf_util.get_dsf_building_df("./2019/"+file, 2019, month))
```

```python
df_2019[df_2019['Month'] == 1].describe()
```

```python
each_month_2019_volume = df_2019.groupby('Month').agg({'Volume': 'sum'})
each_month_2019_volume
```

```python
ax = each_month_2019_volume.plot(figsize=(5,5))
ax.set_xlabel('月份')
ax.set_ylabel('成交量')
ax.set_title('2019年 住宅成交量')
```
